```python
from PIL import Image
import pytesseract

```

## CART ID 139103

```python
#139103
image_path = "/Users/joshuarivera/Desktop/CART_seq_images/139103/Screenshot 2025-01-07 at 4.09.47 PM.png"
image = Image.open(image_path)

extracted_text = pytesseract.image_to_string(image)

extracted_text

new_image_path = "/Users/joshuarivera/Desktop/CART_seq_images/139103/Screenshot 2025-01-07 at 4.09.58 PM.png"
new_image = Image.open(new_image_path)

new_extracted_text = pytesseract.image_to_string(new_image)

combined_text = extracted_text + new_extracted_text

print(combined_text)
```

```python
cleaned_text = combined_text.replace('\n','')

len_cleaned_text= len(cleaned_text)
print(len_cleaned_text)
```
